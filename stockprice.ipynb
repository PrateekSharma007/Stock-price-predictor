```python
sp500 = sp500.history(period = "max")  #history of data 
```

```python
#data cleaning 
```

```python
del sp500["Dividends"] 
del sp500["Stock Splits"]
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1) 
```

```python
sp500
```

```python
sp500["Target"] = (sp500["Tomorrow"]  > sp500["Close"]).astype(int)
```

```python
sp500
```

```python
sp500 = sp500.loc["1990-01-01" :].copy() #clear data before 1990
```

```python
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier
#random forest picks up non linear relation
```

```python
model = RandomForestClassifier(n_estimators=100,min_samples_split=100,random_state=1 )
```

```python
train = sp500.iloc[:-100]
test = sp500.iloc[-100:] 

predictors = ["Close" , "Volume" , "Open" , "High" , "Low"] 
model.fit(train[predictors] , train["Target"])  #train on predictor and predict the values on target values
```

```python
from sklearn.metrics import precision_score #check the accuracy 

preds = model.predict(test[predictors]) 
```

```python
import pandas as pd 
preds = pd.Series(preds,index = test.index)
```

```python
preds 
```

```python
precision_score(test["Target"] , preds)
```

```python
#build a backtesting system 
def predict(train, test, predict, model) : 
    model.fit(train[predictors],train["Target"]) 
    preds = model.predict(test[predictors]) 
    preds = pd.Series(preds,index = test.index ,name = "Predictions")
    combined = pd.concat([test["Target"],preds] ,axis = 1)
    return combined
    

```

```python
def backtesting(data,model,predictors,start = 250 , step =250): 
    all_predictions = [] 

    for i in range(start,data.shape[0] , step) : 
        train = data.iloc[0:i].copy() 
        test = data.iloc[i:(i+step)].copy() 
        predictions = predict(train,test,predictors,model) 
        all_predictions.append(predictions) 
    return pd.concat(all_predictions)


```

```python
predictions = backtesting(sp500,model,predictors) 
```
